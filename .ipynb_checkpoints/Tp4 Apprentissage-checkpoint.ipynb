## Problème jouet

```python
FirstGroup = np.random.normal(loc=1,scale=0.3,size=300)
YFirstGroup = np.random.normal(loc=1,scale=0.3,size=300)
SecondGroup = np.random.normal(loc=-1,scale=0.3,size=300)
YSecondGroup = np.random.normal(loc=-1,scale=0.3,size=300)
combined=np.concatenate((FirstGroup, SecondGroup))
Ycombined=np.concatenate((YFirstGroup, YSecondGroup))
OurDataFrame=pd.DataFrame(combined,columns=['X'])
OurDataFrame['Y']=Ycombined.tolist()
Ones=np.ones(300)
NegOnes=-np.ones(300)
Merged=np.concatenate((Ones,NegOnes))
OurDataFrame['labels']=Merged.tolist()
OurDataFrame
```

```python
x1 = OurDataFrame['X']
x2 = OurDataFrame['Y']
y = OurDataFrame['labels']
plt.scatter(x1[y==-1], x2[y==-1], c='r')
plt.scatter(x1[y==1], x2[y==1], c='b')
```

les cellules ci-dessus sont conçues pour créer un ensemble de données de 600 points où 300 d'entre eux ont une moyenne de (1,1) et une étiquette 1 et 300 d'entre eux ont une moyenne de (-1,-1) et une étiquette -1. chaque classe de points a une variance de 0.3

```python
X_train, X_test, y_train, y_test = train_test_split(OurDataFrame[['X','Y']], OurDataFrame['labels'], test_size=0.3, random_state=42)
X=OurDataFrame[['X','Y']]
y=OurDataFrame['labels']
param_grid = {'C': [10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['linear']}
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
grid.fit(X,y)
```

```python
print(grid.best_params_)
```

pour choisir les meilleurs paramètres, j'ai décidé de les choisir sur la base des résultats de la validation croisée et de la recherche sur grille, donc j'ai obtenu ces paramètres pour les utiliser.

```python
ourmodel=SVC(C =10,gamma=1 ,random_state=0,kernel='linear')
ourmodel.fit(X,y)
grid_predictions = ourmodel.predict(X)
print(classification_report(y, grid_predictions))
```

nous pouvons voir ici, sur la base du rapport de classification, que nous avons une précision de 100%, donc tous nos points sont correctement classifiés.

```python
def calculate_tpr_fpr(y_real, y_pred):
    cm = confusion_matrix(y_real, y_pred)
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]
    TP = cm[1, 1]
    tpr =  TP/(TP + FN) # sensitivity - true positive rate
    fpr = 1 - TN/(TN+FP) # 1-specificity - false positive rate
    
    return tpr, fpr
```

Cette fonction calculera la sensibilité (taux de vrais positifs) et la spécificité (taux de faux positifs).

```python
def get_all_roc_coordinates(y_real, y_proba):    
    tpr_list = [0]
    fpr_list = [0]
    for i in range(len(y_proba)):
        threshold = y_proba[i]
        y_pred = y_proba >= threshold
        tpr, fpr = calculate_tpr_fpr(y_real, y_pred)
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    return tpr_list, fpr_list
```

Calcule toutes les coordonnées de la courbe ROC (tpr et fpr) en considérant chaque point comme un seuil pour la prédiction de la classe.

```python
def plot_roc_curve(tpr, fpr, scatter = True, ax = None):
    if ax == None:
        plt.figure(figsize = (5, 5))
        ax = plt.axes()
    if scatter:
        sns.scatterplot(x = fpr, y = tpr, ax = ax)
    sns.lineplot(x = fpr, y = tpr, ax = ax)
    sns.lineplot(x = [0, 1], y = [0, 1], color = 'green', ax = ax)
    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
```

```python
from sklearn.metrics import roc_auc_score
y_proba=model.predict_proba(X028_test)
plt.figure(figsize = (12, 8))
bins = [i/20 for i in range(20)] + [1]
classes =[0,2,8]
roc_auc_ovr = {}
for i in range(len(classes)):
    c = classes[i]
    df_aux = pd.DataFrame(X028_test.copy())
    df_aux['class'] = [1 if y == c else 0 for y in y028_test]
    df_aux['prob'] = y_proba[:, i]
    df_aux = df_aux.reset_index(drop = True)
    ax = plt.subplot(2, 3, i+1)
    sns.histplot(x = "prob", data = df_aux, hue = 'class', color = 'b', ax = ax, bins = bins)
    ax.set_title(c)
    ax.legend([f"Class: {c}", "Rest"])
    ax.set_xlabel(f"P(x = {c})")
    ax_bottom = plt.subplot(2, 3, i+4)
    tpr, fpr = get_all_roc_coordinates(df_aux['class'], df_aux['prob'])
    plot_roc_curve(tpr, fpr, scatter = False, ax = ax_bottom)
    ax_bottom.set_title("ROC Curve OvR")
    roc_auc_ovr[c] = roc_auc_score(df_aux['class'], df_aux['prob'])
plt.tight_layout()
```

```python
avg_roc_auc = 0
i = 0
for k in roc_auc_ovr:
    avg_roc_auc += roc_auc_ovr[k]
    i += 1
    print(f"{k} ROC AUC: {roc_auc_ovr[k]:.4f}")
print(f"average ROC AUC: {avg_roc_auc/i:.4f}")
```

Ici j'ai utilisé deux méthodes pour mesurer la performance de mon modèle. la première qui était le rapport de classification, et j'ai regardé dans celui-ci la précision sur chaque classe, et j'ai obtenu de très bonnes valeurs (97% pour le 0 , 98% pour le 2 et 95% pour le 8). Ensuite j'ai lu sur la courbe ROC AUC qui trace le taux de vrais positifs de mon algo pour chaque classe, et quand l'aire sous la courbe devient de plus en plus grande, cela signifie que je distingue correctement cette classe des autres, et on peut voir que l'aire sous les 3 courbes est approximativement l'aire optimale. Et enfin, j'ai imprimé le score ROCAUC pour chaque classe, et la moyenne entre eux, et en conclusion, je suis capable de distinguer mes 3 classes avec un pourcentage de 99,64%, ce qui est une très bonne valeur pour un problème de classification multi-classes.
